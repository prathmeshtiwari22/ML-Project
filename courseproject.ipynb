```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset (assuming your dataset is in a DataFrame)
data = pd.read_csv('fashionprathmesh.csv')

# Separate features (X) and the target variable (y)
x = data.drop(columns=['uses_ad_boosts'])
y = data['uses_ad_boosts']
```

```python
df = pd.DataFrame(data)
```

```python
#decision tree
X = df[['price', 'units_sold']]
y = df['uses_ad_boosts']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
print(classifier.predict(sc.transform([[10,5000]])))
```

```python
y_pred = classifier.predict(X_test)
print(y_pred)
print(y_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load your dataset
data = pd.read_csv('fashionprathmesh.csv')

# Select numerical features for clustering
X = data[['units_sold', 'rating', 'product_quality']]

# Specify the number of clusters you want
n_clusters = 3  # You can change this number as needed

# Create a K-Means clustering model
kmeans = KMeans(n_clusters=n_clusters)

# Fit the model to your data
kmeans.fit(X)

# Add cluster labels to your dataset
data['cluster'] = kmeans.labels_

# Visualize the clusters, you can choose which columns to plot
plt.scatter(data['units_sold'], data['rating'], c=data['cluster'], cmap='viridis')
plt.title('K-Means Clustering')
plt.xlabel('Units Sold')
plt.ylabel('Rating')
plt.show()
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

# Load your data from a CSV file
data = pd.read_csv("fashionprathmesh.csv")

# Define the input features (X) and the target variable (y)
X = data[["price", "retail_price"]]
y = data["uses_ad_boosts"]

# Create a Decision Tree classifier (you can choose other algorithms as well)
model = DecisionTreeClassifier()

# Train the model on the entire dataset
model.fit(X, y)

# Now you can use the trained model to predict whether to use ad boost for new data
new_data = {
    "price": [12.0],
    "retail_price": [15]
}

ad_boost_prediction = model.predict(pd.DataFrame(new_data))

# ad_boost_prediction will be 1 (use ad boost) or 0 (don't use ad boost)
print("Ad Boost Prediction:", ad_boost_prediction[0])

```
